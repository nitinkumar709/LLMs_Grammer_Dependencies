```python

# HEAD IMPORTANCE / ABLATION STUDY


!pip -q install transformers torch conllu requests seaborn matplotlib pandas numpy tqdm


import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests

from tqdm import tqdm
from conllu import parse
from transformers import BertTokenizer, BertModel

sns.set(style="whitegrid")

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
MODEL_NAME = "bert-base-uncased"
# CRITICAL: Make sure these lines exist right after loading the model
model = BertModel.from_pretrained(MODEL_NAME, output_attentions=True)
model.to(device) # <--- This sends the "brain" to the RTX 3050
model.eval()




DATASET_URL = "https://raw.githubusercontent.com/UniversalDependencies/UD_English-EWT/master/en_ewt-ud-train.conllu"

NUM_SENTENCES = 300      # keep 300 first, can increase later
MAX_LEN_WORDS = 40


tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)

model = BertModel.from_pretrained(
    MODEL_NAME,
    output_attentions=True
)

model.to(device)
model.eval()


print("Downloading dataset...")

data = requests.get(DATASET_URL).text
sentences = parse(data)

print("Total sentences:", len(sentences))


def tokenize_with_alignment(words):

    mapping = {}
    pos = 1

    for i, w in enumerate(words, start=1):

        subs = tokenizer.tokenize(w)

        if len(subs) == 0:
            subs = ["[UNK]"]

        mapping[i] = list(range(pos, pos + len(subs)))
        pos += len(subs)

    return mapping



def get_attention(words):
    sent = " ".join(words)
    inputs = tokenizer(sent, return_tensors="pt", truncation=True, max_length=128)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        with torch.cuda.amp.autocast(): # This line enables the speed boost
            outputs = model(**inputs)

    return outputs.attentions


def token_to_word(token_matrix, mapping, n_words):

    wm = np.zeros((n_words+1, n_words+1))

    for child in range(1, n_words+1):

        ctoks = mapping[child]

        for parent in range(1, n_words+1):

            ptoks = mapping[parent]

            vals = []

            for c in ctoks:
                for p in ptoks:
                    vals.append(token_matrix[c][p])

            wm[child][parent] = np.mean(vals)

    return wm


def top3_score(word_matrix, heads):

    correct = 0
    total = 0

    n = len(heads)

    for child in range(1, n+1):

        gold = heads[child-1]

        if gold == 0:
            continue

        row = word_matrix[child][1:]

        top3 = np.argsort(row)[-3:] + 1

        if gold in top3:
            correct += 1

        total += 1

    if total == 0:
        return None

    return correct / total


dataset = []

for i in range(min(NUM_SENTENCES, len(sentences))):

    try:
        sent = sentences[i]

        words = []
        heads = []

        for tok in sent:
            if isinstance(tok["id"], int):
                words.append(tok["form"])
                heads.append(tok["head"])

        if len(words) < 3:
            continue

        if len(words) > MAX_LEN_WORDS:
            continue

        dataset.append((words, heads))

    except:
        continue

print("Usable sentences:", len(dataset))


print("\nComputing baseline score...")

baseline_scores = []

for words, heads in tqdm(dataset):

    attentions = get_attention(words)
    mapping = tokenize_with_alignment(words)
    n_words = len(words)

    head_scores = []

    for layer in range(12):
        for head in range(12):

            token_matrix = attentions[layer][0, head].cpu().numpy()

            wm = token_to_word(token_matrix, mapping, n_words)

            s = top3_score(wm, heads)

            if s is not None:
                head_scores.append(s)

    baseline_scores.append(max(head_scores))

BASELINE = np.mean(baseline_scores)

print("Baseline Score:", BASELINE)


results = []

print("\nRunning ablation study...")

for mask_layer in range(12):

    for mask_head in range(12):

        scores = []

        for words, heads in tqdm(dataset, leave=False):

            attentions = get_attention(words)
            mapping = tokenize_with_alignment(words)
            n_words = len(words)

            head_scores = []

            for layer in range(12):
                for head in range(12):

                    # skip masked head
                    if layer == mask_layer and head == mask_head:
                        continue

                    token_matrix = attentions[layer][0, head].cpu().numpy()

                    wm = token_to_word(
                        token_matrix,
                        mapping,
                        n_words
                    )

                    s = top3_score(wm, heads)

                    if s is not None:
                        head_scores.append(s)

            if len(head_scores) > 0:
                scores.append(max(head_scores))

        new_score = np.mean(scores)

        drop = BASELINE - new_score

        results.append([
            mask_layer + 1,
            mask_head + 1,
            new_score,
            drop
        ])

        print(
            f"Layer {mask_layer+1} Head {mask_head+1} "
            f"Drop={drop:.4f}"
        )


df = pd.DataFrame(
    results,
    columns=[
        "Layer",
        "Head",
        "ScoreAfterMask",
        "Drop"
    ]
)

df.to_csv("head_ablation_results.csv", index=False)

print("\nSaved: head_ablation_results.csv")


print("\n==============================")
print("TOP IMPORTANT HEADS")
print("==============================")

print(
    df.sort_values(
        "Drop",
        ascending=False
    ).head(15)
)


pivot = df.pivot(
    index="Layer",
    columns="Head",
    values="Drop"
)

plt.figure(figsize=(12,8))

sns.heatmap(
    pivot,
    annot=True,
    fmt=".3f",
    cmap="Reds"
)

plt.title("Head Importance (Score Drop After Masking)")
plt.xlabel("Head")
plt.ylabel("Layer")

plt.show()


top10 = df.sort_values(
    "Drop",
    ascending=False
).head(10)

labels = [
    f"L{r.Layer}-H{r.Head}"
    for _, r in top10.iterrows()
]

plt.figure(figsize=(10,6))

plt.bar(labels, top10["Drop"])

plt.title("Top 10 Important Heads")
plt.xlabel("Head")
plt.ylabel("Score Drop")

plt.xticks(rotation=45)

plt.show()
```
